```python
# Gauss-Jacobi con diferentes x0
A = np.array([[1, 1], [-2, 5]])
b = np.array([7, 0])

estimaciones = [
    [0, 0],
    [5, 2],
    [10, 5],
    [20, 15],
]

tol = 1e-5
max_iter = 10

plt.figure(figsize=(10, 8))
colors = ['blue', 'red', 'green', 'orange']


for idx, x0 in enumerate(estimaciones):
    x_sol, trajectory = gauss_jacobi(A=A, b=b, x=x0, tol=tol, max_iter=max_iter)
    
    # Convertir trayectoria a array 2D para graficar
    traj = np.array([t.flatten() for t in trajectory])
    
    # Graficar trayectoria
    plt.plot(traj[:, 0], traj[:, 1], 'o-', color=colors[idx], 
             markersize=8, linewidth=2, label=f'x0={x0}')
    
    # Etiquetar punto inicial y final
    plt.annotate(f'0', (traj[0, 0], traj[0, 1]), textcoords="offset points", 
                xytext=(5, 5), ha='center', fontsize=9, fontweight='bold', 
                color=colors[idx])
    plt.annotate(f'{len(traj)-1}', (traj[-1, 0], traj[-1, 1]), textcoords="offset points", 
                xytext=(5, 5), ha='center', fontsize=9, fontweight='bold', 
                color=colors[idx])

plt.xlabel('x1', fontsize=12)
plt.ylabel('x2', fontsize=12)
plt.title('Trayectorias: Gauss-Jacobi (Diferentes x0)', fontsize=14, fontweight='bold')
plt.legend(loc='best')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## Ejercicio 3

```python
# Probar diferentes posiciones iniciales
A = np.array([[1, -1, 2, 0],
              [-1, 1, -1, 3],
              [2, -1, 1, -1],
              [0, 3, -1, 2]])
b = np.array([1360, 1130, 1350, 3650])

# Probar 3 posiciones iniciales diferentes
estimaciones = [
    [0, 0, 0, 0],           
    [100, 100, 100, 100],   
    [1000, 1000, 1000, 1000]
]

tol = 1e-5
max_iter = 50

print("="*70)
print("GAUSS-JACOBI")
print("="*70)
for idx, x0 in enumerate(estimaciones, 1):
    print(f"\nPosición inicial {idx}: {x0}")
    try:
        x_sol, trajectory = gauss_jacobi(A=A, b=b, x=x0, tol=tol, max_iter=max_iter)
        print(f"Solución encontrada: {x_sol.flatten()}")
        print(f"Número de iteraciones: {len(trajectory)-1}")
        
        # Verificar si diverge (valores muy grandes)
        if np.any(np.abs(x_sol) > 1e10):
            print("DIVERGE - Los valores son muy grandes")
    except Exception as e:
        print(f"ERROR o DIVERGENCIA: {e}")

print("\n" + "="*70)
print("GAUSS-SEIDEL")
print("="*70)
for idx, x0 in enumerate(estimaciones, 1):
    print(f"\nPosición inicial {idx}: {x0}")
    try:
        x_sol, trajectory = gauss_seidel(A=A, b=b, x=x0, tol=tol, max_iter=max_iter)
        print(f"Solución encontrada: {x_sol.flatten()}")
        print(f"Número de iteraciones: {len(trajectory)-1}")
        
        # Verificar si diverge (valores muy grandes)
        if np.any(np.abs(x_sol) > 1e10):
            print("DIVERGE - Los valores son muy grandes")
    except Exception as e:
        print(f"ERROR o DIVERGENCIA: {e}")
```
